```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_atvi_weekly_det = pd.read_csv("atvi_weekly_return_detailed.csv")
df_spy_weekly_det = pd.read_csv("spy_weekly_return_detailed.csv")
```

```python
df_atvi_weekly_vol = pd.read_csv("atvi_weekly_return_volatility.csv")
df_spy_weekly_vol = pd.read_csv("spy_weekly_return_volatility.csv")
```

```python
df_atvi_weekly_det = df_atvi_weekly_det[df_atvi_weekly_det["Year"] > 2020]
df_spy_weekly_det = df_spy_weekly_det[df_spy_weekly_det["Year"] > 2020]
```

```python
df_atvi_weekly_vol = df_atvi_weekly_vol[df_atvi_weekly_vol["Year"] > 2020]
df_spy_weekly_vol = df_spy_weekly_vol[df_spy_weekly_vol["Year"] > 2020]
print(f"Activision Blizzard -> \n{df_atvi_weekly_vol.head(10)}")
print(f"S&P 500 -> \n{df_spy_weekly_vol.head(10)}")
```

```python
print("Activision Blizzard Shape ->", df_atvi_weekly_vol.shape)
print("Red -> Will not trade\nGreen -> Will Trade\nCounts ->")
print(df_atvi_weekly_vol["Label"].value_counts())
print("\n")
print("S&P 500 Shape ->", df_atvi_weekly_vol.shape)
print("Red -> Will not trade\nGreen -> Will Trade\nCounts ->")
print(df_atvi_weekly_vol["Label"].value_counts())
```

```python
def scatter_mean_ret_vol(df, stocks, year):
    if year == 1:
        ret = df["mean_return"][:52]
        vol = df["volatility"][:52]
        labels = df["Label"][:52]
    elif year == 2:
        ret = df["mean_return"][52:]
        vol = df["volatility"][52:]
        labels = df["Label"][52:]

    plt.figure(figsize = (15, 10))
    plt.scatter(ret, vol, c = labels)
    plt.plot([-1.5, 2], [0.5, 1], color='red')
    plt.xlabel("Mean Returns")
    plt.ylabel("Volatility")
    plt.title(f"Mean Returns Vs Volatility for {stocks}")
    plt.show()
```

```python
scatter_mean_ret_vol(df_atvi_weekly_vol, "Activision Blizzard", 1)
scatter_mean_ret_vol(df_spy_weekly_vol, "S&P 500", 1)
```

```python
scatter_mean_ret_vol(df_atvi_weekly_vol, "Activision Blizzard", 2)
scatter_mean_ret_vol(df_spy_weekly_vol, "S&P 500", 2)
```
