```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_atvi = pd.read_csv("atvi_weekly_return_detailed.csv")
df_spy = pd.read_csv("spy_weekly_return_detailed.csv")
```

```python
df_atvi["Week-Month"] = df_atvi["Week_Number"] % 4 + 1
df_spy["Week-Month"] = df_spy["Week_Number"] % 4 + 1
```

```python
df_atvi.head()
```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
```

```python
def get_trade_months(df):
    current_month_index = 0
    previous_week = 0
    for index, row in df.iterrows():
        current_week = row['Week-Month']
        if current_week == 1 and previous_week == 4:
            current_month_index += 1
        df.at[index, 'Trade Month'] = months[current_month_index % len(months)]
        previous_week = current_week


get_trade_months(df_atvi)
get_trade_months(df_spy)
```

```python
df_atvi["True Labels"] = np.where(df_atvi["mean_return"] < 0, "red", "green")
df_spy["True Labels"] = np.where(df_spy["mean_return"] < 0, "red", "green")
```

```python
def oracle_predictions(df, P):
    import random
    for i in range(df.shape[0]):
        if random.random() <= P:
            df.at[i, f"Oracle Predictions for P = {P}"] = df.at[i, "True Labels"]
        else:
            if df.at[i, "True Labels"] == "green":
                df.at[i, f"Oracle Predictions for P = {P}"] = "red"
            else:
                df.at[i, f"Oracle Predictions for P = {P}"] = "green"
```

```python
p_values = np.arange(0, 11)/10
for i in p_values:
    oracle_predictions(df_atvi, i)

for i in p_values:
    oracle_predictions(df_spy, i)
```

```python
df_spy_weekly = df_spy.drop_duplicates(subset=["mean_return", "Week-Month", "Trade Month"])
df_spy_weekly.head()
```

```python
df_atvi_weekly = df_atvi.drop_duplicates(subset=["mean_return", "Week-Month", "Trade Month"])
df_atvi_weekly.head()
```

```python
def get_p_1_amount(df, stocks):
    amount = 100
    my_amount = amount + 78
    c = 0
    for i in range(df.shape[0]):
        if df.iloc[i, -1] == "green":
            amount = amount*(1 + (df.iloc[i, 12]/100))
            if amount >= my_amount and c == 0:
                wk = i
                c += 1
    print(f"Week taken to get to my amount for {stocks} -> {wk} weeks")
    print(f"Final Amount for {stocks} -> $ {round(amount, 2)}")


get_p_1_amount(df_atvi_weekly, "Activision Blizzard")
get_p_1_amount(df_spy_weekly, "S&P 500")
```

```python
def get_buy_and_hold_amount(df, stocks):
    amount = 100
    for i in range(df.shape[0]):
        amount = amount*(1 + (df.iloc[i, 12]/100))
    print(f"Final Amount for {stocks} -> $ {round(amount, 2)}")


get_buy_and_hold_amount(df_atvi_weekly, "Activision Blizzard")
get_buy_and_hold_amount(df_spy_weekly, "S&P 500")
```

```python
def get_p_final_val(p, df, stocks):
    labels = np.array(df[f"Oracle Predictions for P = {p}"])
    weekly_returns = np.array(df["mean_return"])
    amount = 100
    for i in range(labels.shape[0]):
        if labels[i] == "green":
            amount = amount*(1 + weekly_returns[i]/100)
    print(f"Final Amount for {p} for {stocks} -> $ {round(amount, 2)}")

p_values = np.arange(0, 11)/10
for i in p_values:
    get_p_final_val(i, df_atvi_weekly, "Activision Blizzard")
    get_p_final_val(i, df_spy_weekly, "S&P 500")
    print("\n")
```

```python
#Scenario 1
def angry_oracle_scenario1(df, stocks):
    best_ten_weeks = df.sort_values(by = "mean_return")[-10:]
    df["Angry Oracle Scenario 1"] = df["Oracle Predictions for P = 1.0"]
    for i in range(10):
        for j in range(df.shape[0]):
            if df.iloc[j, 12] == best_ten_weeks.iloc[i, 12]:
                df.iloc[j, -1] = "red"
    labels = np.array(df[f"Angry Oracle Scenario 1"])
    weekly_returns = np.array(df["mean_return"])
    amount = 100
    for i in range(labels.shape[0]):
        if labels[i] == "green":
            amount = amount*(1 + weekly_returns[i]/100)
    print(f"Angry Oracle Scenario 1 final amount for {stocks} -> $ {round(amount, 2)}")

angry_oracle_scenario1(df_atvi_weekly, "Activision Blizzard")
angry_oracle_scenario1(df_spy_weekly, "S&P 500")
```

```python
#Scenario 2
def angry_oracle_scenario2(df, stocks):
    worst_ten_weeks = df.sort_values(by = "mean_return")[:10]
    df["Angry Oracle Scenario 2"] = df["Oracle Predictions for P = 1.0"]
    for i in range(10):
        for j in range(df.shape[0]):
            if df.iloc[j, 12] == worst_ten_weeks.iloc[i, 12]:
                df.iloc[j, -1] = "green"
    labels = np.array(df[f"Angry Oracle Scenario 2"])
    weekly_returns = np.array(df["mean_return"])
    amount = 100
    for i in range(labels.shape[0]):
        if labels[i] == "green":
            amount = amount*(1 + weekly_returns[i]/100)
    print(f"Angry Oracle Scenario 2 final amount for {stocks} -> $ {round(amount, 2)}")

angry_oracle_scenario2(df_atvi_weekly, "Activision Blizzard")
angry_oracle_scenario2(df_spy_weekly, "S&P 500")
```

```python
#Scenario 3
def angry_oracle_scenario3(df, stocks):
    worst_five_weeks = df.sort_values(by = "mean_return")[:5]
    best_five_weeks = df.sort_values(by = "mean_return")[-5:]
    df["Angry Oracle Scenario 3"] = df["Oracle Predictions for P = 1.0"]
    for i in range(5):
        for j in range(df.shape[0]):
            if df.iloc[j, 12] == worst_five_weeks.iloc[i, 12]:
                df.iloc[j, -1] = "green"
    for i in range(5):
        for j in range(df.shape[0]):
            if df.iloc[j, 12] == best_five_weeks.iloc[i, 12]:
                df.iloc[j, -1] = "red"
    labels = np.array(df[f"Angry Oracle Scenario 3"])
    weekly_returns = np.array(df["mean_return"])
    amount = 100
    for i in range(labels.shape[0]):
        if labels[i] == "green":
            amount = amount*(1 + weekly_returns[i]/100)
    print(f"Angry Oracle Scenario 3 final amount for {stocks} -> $ {round(amount, 2)}")

angry_oracle_scenario3(df_atvi_weekly, "Activision Blizzard")
angry_oracle_scenario3(df_spy_weekly, "S&P 500")
```
